## 2.1 Solucion cerrada de regresion lineal

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
x = np.random.uniform(0, 3, 1000)
y = 5 + 2 * x + np.random.uniform(2, 4, 1000)

# Grfico de dispersin
plt.scatter(x, y, alpha=0.5)
plt.title('Dispersin de datos')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.show()
```

```python
# Funcin para visualizar datos
def visualizar_datos(x, y, n=1000):
    plt.scatter(x, y, alpha=0.5)
    plt.title('Dispersin de datos simulados')
    plt.xlabel('Variable Independiente')
    plt.ylabel('Variable Dependiente')
    plt.show()
```

```python
# Ajuste de regresin lineal
def regresion_lineal(x, y):
    X = np.vstack([np.ones(len(x)), x]).T
    theta = np.linalg.inv(X.T @ X) @ X.T @ y
    return theta
```

```python
# Graficar regresin lineal
theta = regresion_lineal(x, y)
plt.scatter(x, y, alpha=0.5)
plt.plot(x, theta[0] + theta[1] * x, color='red')
plt.show()
```

```python
# Clase regressionLineal
class regressionLineal:
    def __init__(self, tipo_datos='simulado', tamano=1000):
        self.tipo_datos = tipo_datos
        self.tamano = tamano
    
    def visualizar_datos(self, x, y):
        visualizar_datos(x, y)

    def ajustar_regresion(self, x, y):
        return regresion_lineal(x, y)
    
```

## 2.2 Regresion Polinomial

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Cargar datos
data = pd.read_csv('measurements.csv')
x_poly = data['distance']
y_poly = data['consume']

# Visualizacin inicial
plt.scatter(x_poly, y_poly)
plt.xlabel('Distance')
plt.ylabel('Consume')
plt.show()

```

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Regresin polinomial
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(x_poly)
model = LinearRegression().fit(X_poly, y_poly)

# Graficar resultados
plt.scatter(x_poly, y_poly, color='blue')
plt.plot(x_poly, model.predict(X_poly), color='red')
plt.xlabel('Distance')
plt.ylabel('Consume')
plt.show()
```
